```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
pd.options.display.max_columns = 999
import myLogFun
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
import lightgbm as lgb
from sklearn.preprocessing import PolynomialFeatures
```

```python
clustered_df = pd.read_csv('../files/cluster_df.csv')
coor_df = clustered_df[['Latitude','Longitude']]
clustered_df = clustered_df.drop(['Latitude','Longitude','color'],axis=1)
clustered_df_log,log_tran_col = myLogFun.log_tran(clustered_df.copy(),['Total Traffic Volume (MB)','cluster_id'])
clustered_df_log.head()
```

```python
clustered_df_log.describe()
```

```python
clustered_id = pd.get_dummies(clustered_df_log[['cluster_id']],columns=['cluster_id'],drop_first=True)
clustered_df_log.drop('cluster_id',axis=1,inplace=True)
```

```python
X_num = clustered_df_log.drop('Total Traffic Volume (MB)',axis=1).values
y = clustered_df_log['Total Traffic Volume (MB)'].values
scy = StandardScaler()
y = scy.fit_transform(y.reshape(-1,1))
X_cat = clustered_id.values
```

```python
model_lgb = lgb.LGBMRegressor(learning_rate=0.05,n_estimators=2000,num_leaves=100)
```

baseline model

```python
# split train test data
scX = StandardScaler()
X_num = scX.fit_transform(X_num)
print('X_num.shape:',X_num.shape)
print('X_cat.shape:',X_cat.shape)

#concat the categorical X back
X = np.concatenate((X_num,X_cat),axis=1)
print('X.shape:',X.shape)

# split data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
model_lgb.fit(X_train,y_train)
y_pred = model_lgb.predict(X_test)
r2_score(y_test,y_pred)
```

model with polynomial feature

```python
# make polynomial features
poly_maker = PolynomialFeatures(2)
X_poly_num = poly_maker.fit_transform(X_num)

# scale the feature
scX_poly = StandardScaler()
X_poly_num = scX_poly.fit_transform(X_poly_num)

# concat with categorical feature
X_poly = np.concatenate((X_poly_num,X_cat),axis=1)

# split data
X_poly_train,X_poly_test,y_poly_train,y_poly_test=train_test_split(X_poly,y,test_size=0.3, random_state=42)
```

```python
model_lgb.fit(X_poly_train,y_poly_train)
y_poly_pred = model_lgb.predict(X_poly_test)
r2_score(y_poly_test,y_poly_pred)
```

polynomial with only interaction term

```python
# make polynomial features
poly_maker = PolynomialFeatures(2,interaction_only=True)
X_poly_num = poly_maker.fit_transform(X_num)[:,1:]

# scale the feature
scX_poly = StandardScaler()
X_poly_num = scX_poly.fit_transform(X_poly_num)

# concat with categorical feature
X_poly = np.concatenate((X_poly_num,X_cat),axis=1)

# split data
X_poly_train,X_poly_test,y_poly_train,y_poly_test=train_test_split(X_poly,y,test_size=0.3, random_state=42)
```

```python
model_lgb.fit(X_poly_train,y_poly_train)
y_poly_pred = model_lgb.predict(X_poly_test)
r2_score(y_poly_test,y_poly_pred)
```

feature correlation

```python
corr = []
for i in range(X_num.shape[1]):
    corr.append(np.corrcoef(X_num[:,i],X_num[:,8])[0][1])
```

```python
corr = np.array(corr)
```

making prediction of coverage improvement for one grid

```python
target_rsrp = -80
rsrp_arr = np.zeros((10,))
rsrp_arr[8] = target_rsrp
rsrp_arr_tran = scX.transform(rsrp_arr.reshape(1,-1))[0]

index = np.random.choice(range(X_test.shape[0]))
num_feat = X_test[index,:10]
cat_feat = X_test[index,10:]

diff_sc = rsrp_arr_tran[8] - num_feat[8]

new_num_feat = num_feat + diff_sc*corr

feat = np.concatenate((new_num_feat,cat_feat))
pred_traff = model_lgb.predict(feat.reshape(1,-1))
print('current traffic: ',scy.inverse_transform(y_test[index])[0])
print('current rsrp: ',scX.inverse_transform(num_feat)[8])
print('improved rsrp: ',target_rsrp)
print('predicted traffic: ',scy.inverse_transform(pred_traff)[0])
```

# Bring in the PA

```python
import fiona
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
pa = fiona.open("../files/shape files/RNC_PA_region.shp")
```

```python
pa_poly = [Polygon(item['geometry']['coordinates'][0]) for item in pa]
```

```python
clustered_df = pd.concat([coor_df,clustered_df],axis=1)
clustered_df.head()
```

```python
from tqdm import tqdm
tqdm.pandas()

def map_coor2cid(x):
    point = Point(x['Longitude'],x['Latitude'])
    for i,pa in enumerate(pa_poly):
        if pa.contains(point):
            return i
        
clustered_df['pa_id'] = clustered_df.progress_apply(map_coor2cid,axis=1)
```

```python
target_rsrp = -80
rsrp_arr = np.zeros((10,))
rsrp_arr[8] = target_rsrp
rsrp_arr_tran = scX.transform(rsrp_arr.reshape(1,-1))[0]

pred_traff=[]
for i in tqdm(range(X.shape[0])):
    
    num_feat = X[i,:10]
    cat_feat = X[i,10:]

    diff_sc = rsrp_arr_tran[8] - num_feat[8]

    new_num_feat = num_feat + diff_sc*corr

    feat = np.concatenate((new_num_feat,cat_feat))
    pred_traff.append(model_lgb.predict(feat.reshape(1,-1)))
    
```

```python
clustered_df['pred_improved_traffic'] = pred_traff
clustered_df['pred_improved_traffic'] = clustered_df['pred_improved_traffic'].apply(lambda x:scy.inverse_transform(x)[0])
clustered_df['pred_improved_traffic'] = clustered_df['pred_improved_traffic'] - clustered_df['Total Traffic Volume (MB)']
```

```python
pa_df = pd.DataFrame(clustered_df.groupby('pa_id')['pred_improved_traffic'].sum()).reset_index()
```

```python
pa_df.head(100)
```

```python
# find each pa center
pa_lng_center = []
pa_lat_center = []
for item in tqdm(pa):
    lng_sum = 0
    lat_sum = 0
    for coor in item['geometry']['coordinates'][0]:
        lng_sum+=coor[0]
        lat_sum+=coor[1]
    
    pa_lng_center.append(lng_sum/len(item['geometry']['coordinates'][0]))
    pa_lat_center.append(lat_sum/len(item['geometry']['coordinates'][0]))
```

```python
pa_df_2 = pd.DataFrame({'pa_id':range(450),'lng':pa_lng_center,'lat':pa_lat_center})
```

```python
pa_df = pd.merge(pa_df,pa_df_2,on='pa_id')
```
